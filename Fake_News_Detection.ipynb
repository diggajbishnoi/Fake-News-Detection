```python
import pandas as pd
import numpy as np
```

##**Data collection & Loading**:-

```python
url="https://raw.githubusercontent.com/diggajbishnoi/Fake-News-Detection/main/Fake.csv"
fake=pd.read_csv(url)
fake.head()
```

```python
url="https://raw.githubusercontent.com/diggajbishnoi/Fake-News-Detection/main/True.csv"
true=pd.read_csv(url)
true.head()
```

## **Data Preprocessing & Cleaning**:-

```python
## Target column for true news:-

true['label']=1
```

```python
## Target column for fake news:-


fake['label']=0
```

```python
true.head()
```

```python
fake.head()
```

```python
## combining both the datasets:-

news=pd.concat([fake,true],axis=0)
```

```python
news.head()
```

```python
news.tail()
```

```python
## checking the null values:-

news.isnull().sum()
```

```python
news.shape
```

```python
## Dropping the unnecessary columns:-

news=news.drop(['title','subject','date'],axis=1)
```

```python
news.head()
```

```python
## reshuffling of the data:-

news=news.sample(frac=1)           ## RESHUFFLING
```

```python
news.head()
```

```python
## reseting the index :-

news.reset_index(inplace=True)
news.head()
```

```python
## droping the 'index' column:-

news.drop(['index'],axis=1, inplace=True)
news.head()
```

## **Feature Extraction**:-

```python
import re     ## regular expression or RegEx
```

```python
def wordopt(text):

  ## Convert into lower case.
  text=text.lower()

  ## Remove URL's:-
  text=re.sub(r'https?://\S+|www\.\S+','',text)

  ## Remove HTML Tags:-
  text=re.sub(r'<.*?>','',text)

  ## Remove punctuations:-
  text=re.sub(r'[^\w\s]','',text)

  ## Remove digits:-
  text=re.sub(r'\d','',text)

  ## Removing newline characters:-
  text=re.sub(r'\n',' ',text)

  return text
```

```python
news['text']=news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)    ## 0.3 is 30% shape size
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```
